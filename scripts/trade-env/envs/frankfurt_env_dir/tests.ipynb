```python
p = env.prices
```

```python
import matplotlib.pyplot as plt 
```

```python
val = 1000 
N = 2000 

holdings = [] 
w = 1/20.0
for i in range( 20 ):
    holdings.append( val*w/p[ i , 60 ] )
    
```

```python
v = 0 
for  i , h in enumerate(holdings) :
    v += h*p[ i , 60]
v
```

```python
v2 = 0
for  i , h in enumerate(holdings) :
    v2 += h*p[ i , N]
v2
```

```python
returns = (v2 - v)/v
returns
```

```python
years = (N - 60)/365
years
```

```python
annualized = ( 1 + returns )**(1/years) - 1 
```

```python
annualized*100
```
